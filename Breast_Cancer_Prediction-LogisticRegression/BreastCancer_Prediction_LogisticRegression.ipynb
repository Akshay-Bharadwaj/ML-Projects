```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bc_data.csv')
data.head(10)
```

```python
data.isna().sum()
```

```python
data.drop(columns=['Unnamed: 32', 'id'], inplace=True)
```

```python
data.info()
```

```python
data['diagnosis'] = [1 if i == 'M' else 0 for i in data['diagnosis']]
```

```python
data['diagnosis']
```

```python
features = data.drop(columns=['diagnosis']).values
label = data['diagnosis'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')

CL = 0.9

best_ts = -1
best_rs = -1

for rs in range(1, 301):
  X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=rs)

  model = LogisticRegression()

  model.fit(X_train, y_train)

  test = model.score(X_test, y_test)
  train = model.score(X_train, y_train)

  if test > train and test > CL:
    print(f"Test score: {test} | Train score: {train} | Random state: {rs}")

  if test > best_ts:
    best_ts = test
    best_rs = rs

print(f"Best test score: {best_ts} Random state: {best_rs}")
```

```python
# Best Test score is 0.99 with Random state of 265
# Let's create a model for this state

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=265)

model = LogisticRegression()

model.fit(X_train, y_train)

test = model.score(X_test, y_test)
train = model.score(X_train, y_train)

print(f"Test score: {np.round(test, 2)} Random state: 265")
```
